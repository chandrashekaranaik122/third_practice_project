# Removing the outliers from dataset with ZSCORE Method

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

importing the Zscore from the stats model and applying the zscore with numpy absolute function

```python
threshold=3
print(np.where(z>3))
```

setting the threshold to 3 because the 99.9% of data is present in the range of 3 standard deviations 

```python
print(z.iloc[0][6])
print(z.iloc[90][6])
```

clearly we can see the std value is more than the 3 , we need to remove the outliers from dataset

```python
df1=df[(z<3).all(axis=1)]
```

```python
print(df1.shape)
print(df.shape)
```

# Modelling

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

since the target variable is contineous value, its a Regression problem

importing the nessasary lybraries for the Regression model

```python
maxAccu=0
maxRs=0

for i in range(0,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
        
print('best R2 score is ',maxAccu*100,'at the random state',maxRs)
```

performing the linerRegression model for the best  Random state 

```python
from sklearn.tree import DecisionTreeRegressor as DTR
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor,ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,BaggingRegressor
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

Importing the All the  lybraries for the Regression model

```python
print('DecisionTreeRegressor')
dtc=DTR()
dtc.fit(x_train,y_train)
preddtc=dtc.predict(x_test)
preddtc_train=dtc.predict(x_train)
print('R2_score on test data',r2_score(y_test,preddtc)*100)
print('R2_score on train data',r2_score(y_train,preddtc_train)*100)
print('Mean absolute error',mean_absolute_error(y_test,preddtc))
print('Mean squared error ',mean_squared_error(y_test,preddtc))
```

with DecisionTreeRegressor R2_score is 84%

```python
print('Suppport vector mission')
svr=SVR()
svr.fit(x_train,y_train)
predsvr=svr.predict(x_test)
predsvr_train=svr.predict(x_train)
print('R2_score on test data',r2_score(y_test,predsvr)*100)
print('R2-score on train data',r2_score(y_train,predsvr_train)*100)
print('Mean absolute error',mean_absolute_error(y_test,predsvr))
print('Mean squared error',mean_squared_error(y_test,predsvr))
```
