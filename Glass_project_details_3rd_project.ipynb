Modelling

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

importing classifictions models which needed for the prediction

```python
maxAccu=0
maxRs=0

for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    dtc=DecisionTreeClassifier()
    dtc.fit(x_train,y_train)
    pred=dtc.predict(x_test)
    accu=accuracy_score(y_test,pred)
    if accu>maxAccu:
        maxAccu=accu
        maxRs=i
print('the Best accuracy score is =',maxAccu,'at the random state of =',maxRs)
```

checking for the best random state for the predction of my model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=maxRs)
```

splitting the model with best accuracry score 

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.svm import SVC
```

importing all the nessasary models for the best prediction for dataset

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print('Accuracy score',accuracy_score(y_test,predrf)*100)
print('classification report','\n',classification_report(y_test,predrf))
print('confusion matrix','\n',confusion_matrix(y_test,predrf))
```

with RandomForestClassifier the accuracy score is 99.20 and f1 score is 1 , precision also 1

```python
et=ExtraTreesClassifier()
et.fit(x_train,y_train)
predet=et.predict(x_test)
print('Accuracy Score',accuracy_score(y_test,predet)*100)
print('Classification report','\n',classification_report(y_test,predet))
print('Confusion matrix ','\n',confusion_matrix(y_test,predet))
```

with ExtraTreesClassifier accuracy score is 99.20 and f1 score is 1 , precision is 1 

```python
bg=BaggingClassifier()
bg.fit(x_train,y_train)
predbg=bg.predict(x_test)
print('Accuracy score',accuracy_score(y_test,predbg)*100)
print('classification report','\n',classification_report(y_test,predbg))
print('confusion matrix','\n',confusion_matrix(y_test,predbg))
```

BaggingClassifier the Accuracy score 100 and all parameters are good with score 1

```python
ad=AdaBoostClassifier()
ad.fit(x_train,y_train)
predad=ad.predict(x_test)
print('Accuracy Score',accuracy_score(y_test,predad)*100)
print('classification report','\n',classification_report(y_test,predad))
print('confusion matrix','\n',confusion_matrix(y_test,predad))
```
